```python
import keras
import os
import warnings
import cv2
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
model = keras.models.load_model('handwritten_model')
```

```python
i = 0
while os.path.isfile(f'digits/dig{i}.png'):
    try: 
        img = cv2.imread(f'digits/dig{i}.png')[:,:,0]
        img = np.invert(np.array([img]))
        pred = model.predict(img)
        print("prediction : ",np.argmax(pred))
        plt.imshow(img[0], cmap = plt.cm.binary); plt.show()
    except: print("err")
    finally: i+=1
```
